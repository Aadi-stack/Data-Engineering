```python
import pandas as pd

# Load datasets
patients = pd.read_csv('/content/patients.csv')
encounters = pd.read_parquet('/content/encounters.parquet')
medications = pd.read_csv('/content/medications.csv')
symptoms = pd.read_csv('/content/symptoms.csv')
conditions = pd.read_csv('/content/conditions.xlsx - Sheet1.csv')

# Function to summarize dataset
def explore_data(df, name):
    print(f"--- {name} ---")
    print("Shape:", df.shape)
    print("\nMissing Values:")
    print(df.isnull().sum())
    print("\nData Types:")
    print(df.dtypes)
    print("\nDuplicates:", df.duplicated().sum())
    print("\nSample Data:")
    print(df.head())
    print("-" * 50)

# Explore datasets
explore_data(patients, "Patients")
explore_data(encounters, "Encounters")
explore_data(medications, "Medications")
explore_data(symptoms, "Symptoms")
explore_data(conditions, "Conditions")
```
