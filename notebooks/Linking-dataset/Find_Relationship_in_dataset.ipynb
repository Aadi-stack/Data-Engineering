```python
from graphviz import Digraph
from IPython.display import Image

# Create a directed graph
dot = Digraph(comment='Healthcare Data Relationships')

# Add nodes for datasets
dot.node('Patients', 'Patients\n(PATIENT_ID, Gender, Race, Ethnicity, Age, Expenses)')
dot.node('Encounters', 'Encounters\n(Id, PATIENT, Encounter Class, Reason Code)')
dot.node('Symptoms', 'Symptoms\n(PATIENT, Gender, Race, Symptom Count)')
dot.node('Medications', 'Medications\n(PATIENT, Encounter ID, Drug, Reason Code)')
dot.node('Conditions', 'Conditions\n(PATIENT, Diagnosis, Severity)')

# Define relationships
dot.edge('Patients', 'Encounters', label='PATIENT_ID ↔ PATIENT')
dot.edge('Encounters', 'Medications', label='Id ↔ ENCOUNTER')
dot.edge('Patients', 'Medications', label='PATIENT_ID ↔ PATIENT')
dot.edge('Encounters', 'Symptoms', label='PATIENT ↔ PATIENT')
dot.edge('Patients', 'Symptoms', label='PATIENT_ID ↔ PATIENT')
dot.edge('Patients', 'Conditions', label='PATIENT_ID ↔ PATIENT')

# Render and visualize
dot.render('healthcare_data_relationships', format='png', cleanup=True)
Image(filename='healthcare_data_relationships.png')
```
